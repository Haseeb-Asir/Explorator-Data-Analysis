```python
# ðŸ“˜ Tips Dataset EDA
# Author: MUHAMMAD HASEEB ASIR
# Date: 2025-05-11


# ðŸ“Œ 1. Import Libraries

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="whitegrid")
# ðŸ“Œ 2. Load the Dataset

df = sns.load_dataset("tips")

# Display first few rows
df.head()
```

```python

# ðŸ“Œ 3. Basic Information

print("Shape of the dataset:", df.shape)
print("\nDataset Info:\n")
df.info()

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# =============================
# ðŸ“Œ 4. Descriptive Statistics
# =============================
df.describe()
```

```python

# ðŸ“Œ 5. Univariate Analysis (Histograms)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['total_bill'], bins=30, color='skyblue')
plt.title('Distribution of Total Bill')

plt.subplot(1, 2, 2)
sns.histplot(df['tip'], bins=30, color='salmon')
plt.title('Distribution of Tip')

plt.tight_layout()
plt.show()
```

```python
# ======================================
# ðŸ“Œ 6. Count Plots for Categorical Data
# ======================================
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.countplot(x='sex', data=df)
plt.title('Count by Sex')

plt.subplot(1, 3, 2)
sns.countplot(x='smoker', data=df)
plt.title('Count by Smoker')

plt.subplot(1, 3, 3)
sns.countplot(x='day', data=df)
plt.title('Count by Day')


plt.show()
```

```python
# =========================
# ðŸ“Œ 7. Boxplots & Outliers
# =========================
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Total Bill Distribution per Day')
plt.show()
```

```python
# ===========================
# ðŸ“Œ 8. Relationship Analysis
# ===========================

# ðŸ’¡ Tip vs Total Bill
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', style='smoker')
plt.title('Tip vs Total Bill')
plt.show()
```

```python
# ===================================
# ðŸ“Œ 10. Pairplot (Multi-variable View)
# ===================================
sns.pairplot(df, hue='sex', palette='husl')
plt.suptitle("Pairplot of Tips Dataset", y=1.02)
plt.show()
```
