```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df1 = df[['Geography','Gender','Exited']]
```

```python
df1
```

```python
sns.countplot(x=df1['Gender'],hue=df1['Exited'])
```

```python
sns.countplot(x = df1['Geography'],hue=df1['Exited'])
```

```python
#univariate


for i in df1.drop(columns='Exited') :
    plt.figure()
    sns.countplot(data=df1,x = i,hue=df1['Exited'])                                                   
```

```python
# BIVARIATE
sns.histplot(x='Gender',hue='Geography',data=df1,stat = "count",multiple ="dodge")
```
